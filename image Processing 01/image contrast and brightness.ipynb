```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('raza.jpg')

# Set the percent of contrast you want, in this case we are inreasing the contras 
Contrast_percent = 80
high_contrast = image * (1 + Contrast_percent / 100)

# Remember that numpy rounds of, so we need to clip values above 255 and below 0 
high_contrast = np.clip(high_contrast, 0, 255) 
high_contrast = np.uint8(high_contrast)

# Alternatively you can also divide by 255 and convert to float.

plt.figure(figsize=[18,18])

plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Normal");

plt.subplot(132);plt.imshow(high_contrast[...,::-1]);plt.title("High Contrast");
```

### Save image 

```python
filename = 'E:/raza/raza.jpg'
cv2.imwrite(filename, high_contrast)
```

## Brightness Enhancement

```python
image = cv2.imread('raza.jpg')

# Set some brightness constant value 
constant = 80

# Add constant for increasing brightness, we convert to uint32 to prevent numpy r 
bright_image = np.int32(image) + constant

# Remember that numpy rounds of, so we need to clip values above 255 and below 0 
bright_image = np.clip(bright_image, 0, 255) 
bright_image = np.uint8(bright_image)

plt.figure(figsize=[18,18])

plt.subplot(131);plt.imshow(image[...,::-1]);plt.title("Normal");plt.axis('off'); plt.subplot(132);plt.imshow(bright_image[...,::-1]);plt.title("High Brightness");
```
