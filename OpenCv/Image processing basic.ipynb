```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cv=cv2.imread('parrot1.jpg',1)
print(cv.shape)
```

```python
%matplotlib inline
plt.imshow(cv)
```

```python
%matplotlib notebook
plt.imshow(cv)
```

```python
# convert image bgr to rgb 
rgb_image=cv2.cvtColor(cv,cv2.COLOR_BGR2RGB)
%matplotlib inline
plt.imshow(rgb_image)
```

```python
# convert image bgr to GRAY
Gray_image=cv2.cvtColor(cv,cv2.COLOR_BGR2GRAY)
%matplotlib inline
plt.imshow(Gray_image,cmap='gray')
```

```python
# using numpy image convert bgr to rgb
%matplotlib inline
plt.imshow(cv[:,:,::-1])
```

```python
# another methord using numpy image convert rgb to bgr
plt.imshow(rgb_image[...,::-1])
```

```python
plt.figure(figsize=[7,7])

plt.subplot(121)
plt.imshow(cv[...,::-1])
plt.title('Color Image')


plt.subplot(122)
plt.imshow(Gray_image,cmap='gray')
plt.title('Gray Image')
```

## Drow a Line

```python
img=rgb_image.copy()
cv2.line(img,(100,500),(200,20),(0,0,255),5)
plt.imshow(img[...,::-1])
```

## Circle

```python
img=rgb_image.copy()
cv2.circle(img,(528,210),100,(0,255,255),15)
%matplotlib inline
plt.imshow(img[...,::-1])
```

## Rectangle

```python
img=rgb_image.copy()
cv2.rectangle(img,(408,83),(669,464),(0,255,255),15)
%matplotlib inline
plt.imshow(img[...,::-1])
```
